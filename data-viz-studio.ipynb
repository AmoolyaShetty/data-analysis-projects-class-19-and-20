```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to 
## get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()
#df.info()
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
keys = list(df[df["year"].notnull()]["year"].unique())
text_reviews_count =[]

# 
for k in keys:
	sum = int(df[df["year"] == k]["text_reviews_count"].sum()) 
	text_reviews_count.append(sum)

plt.bar(x=keys, height=text_reviews_count,width=.4, color ='maroon')

plt.title("Total text reviews count")
plt.xlabel("year")
plt.ylabel("Text reviews count")
plt.show()            
```

```python
# Use your Python skills to create a visualization here.
import matplotlib.pyplot as plt

keys = list(df[df["year"].notnull()]["year"].unique())
text_reviews_count = []

# Calculate the sum of text reviews for each year
for k in keys:
    total = int(df[df["year"] == k]["text_reviews_count"].sum()) 
    text_reviews_count.append(total)

# Create a line plot
plt.plot(keys, text_reviews_count, marker='o', color='maroon', linestyle='-')

# Add titles and labels
plt.title("Total Text Reviews Count by Year")
plt.xlabel("Year")
plt.ylabel("Text Reviews Count")

# Display the plot
plt.show()


```
